```python
import os
import pandas as pd

# 当前在 src/data_helper 里，所以项目根目录是再往上两层
PROJECT_ROOT = os.path.dirname(os.path.dirname(os.getcwd()))
DATA_DIR = os.path.join(PROJECT_ROOT, "data")
PROCESSED_DIR = os.path.join(DATA_DIR, "processed")

snapshot_path = os.path.join(PROCESSED_DIR, "player_snapshot.parquet")
print("CWD:", os.getcwd())
print("snapshot_path:", snapshot_path)
print("exists?", os.path.exists(snapshot_path))

reg1 = pd.read_parquet(snapshot_path)
reg1.head()
```

```python
# Analyze null values
print("=" * 60)
print("NULL VALUE ANALYSIS")
print("=" * 60)

null_counts = reg1.isnull().sum()
null_percentages = (null_counts / len(reg1) * 100).round(2)

null_summary = pd.DataFrame({
    'Column': null_counts.index,
    'Null_Count': null_counts.values,
    'Null_Percentage': null_percentages.values
}).sort_values('Null_Count', ascending=False)

# Show columns with nulls
null_summary = null_summary[null_summary['Null_Count'] > 0]

print(f"\nTotal rows: {len(reg1):,}")
print(f"Columns with nulls: {len(null_summary)}")
print(f"\nTop 20 columns with most nulls:\n")
print(null_summary.head(20).to_string(index=False))

# Categorize nulls
print("\n" + "=" * 60)
print("NULL CATEGORIZATION")
print("=" * 60)

performance_cols = ['minutes_total', 'goals_per_90', 'assists_per_90', 'games_played', 'minutes_per_game']
club_cols = ['club_win_rate', 'club_goal_diff_per_game', 'club_total_market_value']
player_cols = ['age', 'height_in_cm', 'position', 'name']

print("\nPerformance columns nulls:")
for col in performance_cols:
    if col in reg1.columns:
        null_pct = (reg1[col].isnull().sum() / len(reg1) * 100)
        print(f"  {col:25s}: {reg1[col].isnull().sum():6,} ({null_pct:5.2f}%)")

print("\nClub columns nulls:")
for col in club_cols:
    if col in reg1.columns:
        null_pct = (reg1[col].isnull().sum() / len(reg1) * 100)
        print(f"  {col:25s}: {reg1[col].isnull().sum():6,} ({null_pct:5.2f}%)")

print("\nPlayer info columns nulls:")
for col in player_cols:
    if col in reg1.columns:
        null_pct = (reg1[col].isnull().sum() / len(reg1) * 100)
        print(f"  {col:25s}: {reg1[col].isnull().sum():6,} ({null_pct:5.2f}%)")
```
